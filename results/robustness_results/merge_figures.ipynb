```python
from pdf2image import convert_from_path
from PIL import Image

# List your pdf files
pdf_files = [
    "brightness_decrease.pdf",
    "brightness_increase.pdf",
    "contrast_decrease.pdf",
    "contrast_increase.pdf",
    "gaussian_blur.pdf",
    "gaussian_noise.pdf",
    "occlusion_increase.pdf",
    "salt_and_pepper_noise.pdf"
]

# Convert the first (and only) page of each PDF to a PIL image at a given DPI (300 dpi here)
images = [convert_from_path(pdf, dpi=300)[0] for pdf in pdf_files]

# Assume all images are the same size (width, height)
width, height = images[0].size

# Define grid dimensions: 4 columns and 2 rows for 8 images.
columns = 2
rows = 4

# Create a blank canvas for the collage
collage_width = columns * width
collage_height = rows * height
collage = Image.new('RGB', (collage_width, collage_height), color='white')

# Paste each image into the canvas
for idx, img in enumerate(images):
    x = (idx % columns) * width
    y = (idx // columns) * height
    collage.paste(img, (x, y))

# Save the collage as a PDF (this creates a single-page PDF with the merged figures)
collage.save("full_robustness_figure_2_4.pdf", "PDF", resolution=300)
```
